# Get Page Links

This was a first try at determining a site map for Loblaws.ca.  But, the robots.txt file already provided a site map, so this is not needed.

```python
import os
project_dir = os.getcwd()  # Use the current working directory
html_dir = os.path.join(project_dir, 'html')

main_grocery_file = os.path.join(html_dir, 'natural_organic.html')
```

```python
import bs4
with open(main_grocery_file, 'r') as f:
    main_soup = bs4.BeautifulSoup(f, 'html.parser')
```

```python
# Extract all links from the page
all_links = [a['href'] for a in main_soup.find_all('a', href=True)]

# Filter for only links that start /en/food/
filtered_links = [link for link in all_links if link.startswith('/en/food/')]

# Remove the ? and anything after it from the links
filtered_links = [link.split('?')[0] for link in filtered_links]

# Print the filtered links
print(len(filtered_links))

# Remove duplicates
filtered_links = list(set(filtered_links))
```

```python
for link in filtered_links:
    print(link)
```
